Import necessary libraries

```python
import pickle
import numpy as np
from keras.models import Sequential, load_model
from keras.layers import LSTM, GRU
from keras.callbacks import ModelCheckpoint
from keras.optimizers import SGD
import gensim
from sklearn.model_selection import train_test_split
```

Open conversation tokens from preprocessing

```python
with open('conversation.pickle', 'rb') as f:
    vec_x,vec_y=pickle.load(f)    
```

Convert to float and divide into trainset and testset

```python
vec_x=np.array(vec_x,dtype=np.float64)
vec_y=np.array(vec_y,dtype=np.float64)    

x_train,x_test, y_train,y_test = train_test_split(vec_x, vec_y, test_size=0.2, random_state=1)
```

```python
print(x_train.shape)
print(x_test.shape)
```

Define a 4-layered LSTM model

```python
model=Sequential()
model.add(LSTM(units=300,input_shape=x_train.shape[1:],return_sequences=True, kernel_initializer='glorot_uniform', recurrent_initializer='glorot_uniform', activation='sigmoid'))
model.add(LSTM(units=300,input_shape=x_train.shape[1:],return_sequences=True, kernel_initializer='glorot_uniform', recurrent_initializer='glorot_uniform', activation='sigmoid'))
model.add(LSTM(units=300,input_shape=x_train.shape[1:],return_sequences=True, kernel_initializer='glorot_uniform', recurrent_initializer='glorot_uniform', activation='sigmoid'))
model.add(LSTM(units=300,input_shape=x_train.shape[1:],return_sequences=True, kernel_initializer='glorot_uniform', recurrent_initializer='glorot_uniform', activation='sigmoid'))
```

Define location to save the model

```python
filename = './model/model_lstm.h5'

model.compile(loss='cosine_proximity', optimizer='adam', metrics=['accuracy'])
```

Start train the network for 10000 epochs<br>
run again to further train the network

```python
callback = ModelCheckpoint(filename, monitor='val_loss', verbose=0, save_best_only=True)

model.fit(x_train, y_train, batch_size=200, epochs=1000,validation_data=(x_test, y_test), verbose=2, callbacks=[callback])
```

To test the network, load the gensim word2vec model
